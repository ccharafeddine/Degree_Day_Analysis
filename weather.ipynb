```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
import requests
import re
```

```python
executable_path = {'executable_path': '/Users/bigchafo/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://www.wunderground.com/history/daily/us/tx/corpus-christi/KCRP/date/2021-2-28"

browser.visit(url) # Use Splinter to visit the url
```

```python
# Create a html object of the website
html_table = browser.html

# Parse the website HTML with beatifulsoup

scrape_soup = bs(html_table, 'html.parser')

# Print the results
# print(table_weather.prettify())
```

```python
table_weather = scrape_soup.find_all("table", attrs={"class": "mat-table"})
print("Number of tables in the side: ", len(table_weather))

```

```python
# Let's get the first table

table_one = table_weather[0]

# The head will be our columns names
table_body = table_one.find_all("tr")

# Head values (column names) are the first items of the body list
columns = table_body[0] # The 0th item is the header row

table_body_rows = table_body[1:] # all other items becomes the rest of the rows
# table_body_rows

```

```python
# Time to do the loop thru the HTML columns/headings

# empty list first
table_columns = []

for item in columns.find_all("th"): # will loop thru each elements
    item = (item.text).rstrip("\n") #convert elements to text
    table_columns.append(item) # append to empyt list


print(table_columns)
```

```python
# Now lets pull the rows of the body

# print(table_body_rows)
all_rows = [] #empty list

for row_item in range(len(table_body_rows)): # One row at the time
    each_row = [] #holding row entries for one row
    for row_item in table_body_rows[row_item].find_all("td"): #loop all row entries
        remove_tags = re.sub("(\xa0)|(\n),","",row_item.text)
        each_row.append(remove_tags) #append remove tag to each row 
    
    # now append all rows
    all_rows.append(each_row)
    
```

```python
table_df = pd.DataFrame(data=all_rows, columns=table_columns)
table_df
```

```python
# Send the table into a csv file

table_df.to_csv('/Users/bigchafo/Documents/GitHub/Degree_Day_Analysis/Resources/WeatherUnderground/Corpus_Christi/Corpus_Christi_2021_02_28.csv')
```
